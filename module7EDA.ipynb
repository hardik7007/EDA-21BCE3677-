```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Loading the dataset
df = pd.read_csv('/content/lithium-ion batteries.csv')

# Checking the first few rows and columns
print("Dataset Columns:", df.columns)
print("First few rows of the dataset:")
print(df.head())



```

```python
# Selecting features and target variable
X = df[['Density (gm/cc)']]  # Predictor
y = df['Formation Energy (eV)']  # Target variable

# Splitting the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)

# Training the Linear Regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the target values for the test set
y_pred = regressor.predict(X_test)

# Model evaluation
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)


```

```python
# Plotting the actual vs predicted values
evaluate = pd.DataFrame({'Actual': y_test.values.flatten(), 'Predicted': y_pred.flatten()})
evaluate.head(10).plot(kind='bar', figsize=(10,6), title="Actual vs Predicted Formation Energy (eV) BY 21BCE3677")
plt.xlabel("Sample")
plt.ylabel("Formation Energy (eV) BY 21BCE3677")
plt.show()
```

```python
# Plotting the regression line
sns.lmplot(x='Density (gm/cc)', y='Formation Energy (eV)', data=df, height=6, aspect=1.5)
plt.title("Regression Line for Density vs. Formation Energy BY 21BCE3677")
plt.show()
```

```python
# Displaying evaluation metrics
print(f"R2 score: {r2}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
```
